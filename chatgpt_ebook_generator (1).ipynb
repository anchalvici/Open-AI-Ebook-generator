```python
import os
from openai import OpenAI
chatgpt_api_key = #
client = OpenAI(api_key=chatgpt_api_key)
```

```python
input_topic = 'indian food recipes'
input_target = 'teens'
```

```python
outline_prompt = 'generate outline for an ebook about '+input_topic+' targeted at '+input_target+' with 40-50 subsections.Remember to specify chapter headings as chapter x:introduction, chapter y:etc'
```

```python
# Call the completion endpoint (for text generation)
response = client.chat.completions.create(
    model="gpt-3.5-turbo",  # Specify which GPT model to use
    messages= [{"role": "user", "content": outline_prompt}]
)

# Get the generated text
content = response.choices[0].message.content

# Print or use the result
print(content)
```

```python
import re
# Splitting the text using regular expression to find the chapters
chapters = re.split(r'\bChapter \d+:', content)
chapters = [chapter.strip() for chapter in chapters if chapter.strip()]

# Creating a dictionary to store chapters and their contents
chapter_dict = {}

# Iterating over each chapter
for chapter in chapters:
    # Splitting each chapter into title and contents
    lines = chapter.split('\n')
    title = lines[0].strip()
    contents = [line.strip('') for line in lines[1:]]
    chapter_dict[title] = contents
```

```python
chapter_dict
```

```python
# Function to remove symbols from a string
def remove_symbols(string):
    return re.sub(r'[^a-zA-Z0-9\s]', '', string).strip()

# New dictionary to store modified values
modified_dict = {}

# Iterate through the original dictionary
for key, values in chapter_dict.items():
    modified_values = []
    # Iterate through the values in the list
    for value in values:
        # Remove symbols and keep alphanumeric characters
        modified_value = remove_symbols(value)
        modified_values.append(modified_value)
    # Store modified values in the new dictionary
    modified_dict[key] = modified_values
```
